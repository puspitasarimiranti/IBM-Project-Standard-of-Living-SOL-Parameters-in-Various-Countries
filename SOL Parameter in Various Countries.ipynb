```python
import pandas as pd

#Define the file path
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-AI0272EN-SkillsNetwork/labs/dataset/2016.csv"

#Read the CSV file into a pandas DataFrame, assuming the first row as headers
try:
    df = pd.read_csv(file_path)

    # Print the first 5 rows of the DataFrame
    print(df.head())

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
#Check unique value in every column
print('There are', df['Country'].nunique(), 'unique Country Level in the survey:')
print(df['Country'].value_counts())

print('\nThere are', df['Region'].nunique(), 'unique Region values in the survey:')
print(df['Region'].value_counts())

print('There are', df['Happiness Rank'].nunique(), 'unique Happiness Rank in the survey:')
print(df['Happiness Rank'].value_counts())

print('\nThere are', df['Happiness Score'].nunique(), 'unique Happiness Score values in the survey:')
print(df['Happiness Score'].value_counts())

print('There are', df['Lower Confidence Interval'].nunique(), 'unique Lower Confidence Interval in the survey:')
print(df['Lower Confidence Interval'].value_counts())

print('\nThere are', df['Upper Confidence Interval'].nunique(), 'unique Upper Confidence Interval values in the survey:')
print(df['Upper Confidence Interval'].value_counts())

print('There are', df['Economy (GDP per Capita)'].nunique(), 'unique Economy (GDP per Capita) in the survey:')
print(df['Economy (GDP per Capita)'].value_counts())

print('\nThere are', df['Family'].nunique(), 'unique Family values in the survey:')
print(df['Family'].value_counts())

print('\nThere are', df['Health (Life Expectancy)'].nunique(), 'unique Health (Life Expectancy) values in the survey:')
print(df['Health (Life Expectancy)'].value_counts())

print('There are', df['Freedom'].nunique(), 'unique Freedom in the survey:')
print(df['Freedom'].value_counts())

print('\nThere are', df['Trust (Government Corruption)'].nunique(), 'unique Trust (Government Corruption) in the survey:')
print(df['Trust (Government Corruption)'].value_counts())

print('\nThere are', df['Generosity'].nunique(), 'unique Generosity values in the survey:')
print(df['Generosity'].value_counts())

print('There are', df['Dystopia Residual'].nunique(), 'unique Dystopia Residual in the survey:')
print(df['Dystopia Residual'].value_counts())
```

```python
#Find columns with empty strings
columns_with_empty_strings = []
for column in df.columns:
    if (df[column] == ' ').any():
        columns_with_empty_strings.append(column)

print("Columns with empty strings:", columns_with_empty_strings)
```

```python
#Replace empty strings with NaN in all identified columns
import numpy as np
for column in columns_with_empty_strings:
    df[column] = df[column].replace(' ', np.nan)

print("\nDataFrame after cleaning all columns with empty strings:")
print(df)
```

```python
#Count the number of missing values in each column using isnull().sum()
print(df.isnull().sum())
```

```python
#Convert string to float
df['Upper Confidence Interval'] = df['Upper Confidence Interval'].astype(float)
df['Economy (GDP per Capita)'] = df['Economy (GDP per Capita)'].astype(float)
df['Health (Life Expectancy)'] = df['Health (Life Expectancy)'].astype(float)
df['Freedom'] = df['Freedom'].astype(float)

df.info('Freedom')
```

```python
#Lower Confidence Interval
#calculate the mean value for the "Lower Confidence Interval" column
avg_lci = df["Lower Confidence Interval"].astype("float").mean(axis=0)
print("Lower Confidence Interval:", avg_lci)

#replace the missing 'Lower Confidence Interval' values by the mean value
df["Lower Confidence Interval"].replace(np.nan, avg_lci, inplace=True)
```

```python
#Upper Confidence Interval
#calculate the mean value for the "Upper Confidence Interval" column
avg_uci = df["Upper Confidence Interval"].astype("float").mean(axis=0)
print("Upper Confidence Interval:", avg_uci)

#replace the missing 'Upper Confidence Interval' values by the mean value
df["Upper Confidence Interval"].replace(np.nan, avg_uci, inplace=True)
```

```python
#Economy (GDP per Capita)
#calculate the mean value for the "Economy (GDP per Capita)" column
avg_economy = df["Economy (GDP per Capita)"].astype("float").mean(axis=0)
print("Economy (GDP per Capita):", avg_lci)

#replace the missing 'Economy (GDP per Capita)' values by the mean value
df["Economy (GDP per Capita)"].replace(np.nan, avg_economy, inplace=True)
```

```python
#Health (Life Expectancy)
#calculate the mean value for the "Health (Life Expectancy)" column
avg_health = df["Health (Life Expectancy)"].astype("float").mean(axis=0)
print("Health (Life Expectancy):", avg_lci)

#replace the missing 'Health (Life Expectancy)' values by the mean value
df["Health (Life Expectancy)"].replace(np.nan, avg_health, inplace=True)
```

```python
#Freedom
#calculate the mean value for the "Freedom" column
avg_freedom = df["Freedom"].astype("float").mean(axis=0)
print("Freedom:", avg_lci)

#replace the missing 'Freedom' values by the mean value
df["Freedom"].replace(np.nan, avg_freedom, inplace=True)
```
